## 1. Load the dataset

````python
from pathlib import Path
from red2400 import RED2400Reader, by_reason

# Adjust this path to wherever you saved the JSONL file from Zenodo.
DATA_PATH = Path("rejection_outcomes.jsonl")

reader = RED2400Reader(DATA_PATH)
df = reader.load()
df.head()
````

## 2. Summary statistics

````python
stats = reader.describe()
print(f"Rows: {stats['n_rows']}")
print(f"Date range: {stats['date_range']}")
print(f"Unique mints: {stats['unique_mints']}")
print("Top reject reasons:")
for reason, count in sorted(
    stats['reject_reason_counts'].items(),
    key=lambda kv: kv[1],
    reverse=True,
)[:5]:
    print(f"  {reason}: {count}")
````

## 3. Filter by reject reason

````python
ultra_fast = by_reason(df, "ultra_fast_reject")
print(f"{len(ultra_fast)} rows with reason 'ultra_fast_reject'")
ultra_fast[['symbol', 'ageMin', 'priceChange24h', 'liquidity']].describe()
````

## 4. Plot 24-hour price change distribution

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))
df['priceChange24h'].dropna().clip(-100, 200).hist(bins=60, ax=ax)
ax.set_xlabel('priceChange24h (%, clipped to [-100, 200])')
ax.set_ylabel('Count')
ax.set_title('RED-2400: post-rejection 24h price change distribution')
fig.tight_layout()
plt.show()
````
